```
removePunc = udf(lambda z: re.sub('[^\w\s]', ' ', z).lower())
r2o_punc = r2o_costed.withColumn('item_name', removePunc(r2o_costed.item_name))
r2o_punc.createOrReplaceTempView('r2o_punc_view')
spark.sql('SELECT item_name FROM r2o_punc_view WHERE item_name LIKE \'%1%\'').show()
```

```
#r2oc_agg = r2o_costed.groupby("Order ID").agg({'Quantity':'sum','Item cost':'sum'})
#r2oc_agg2 = r2oc_agg.select(round(r2oc_agg["sum(Item cost)"],2))#.alias("Order cost"))

#date_time = now.strftime("%m/%d/%Y, %H:%M:%S") order_id, 
r2oc_agg = spark.sql(
  """SELECT order_id, MIN(total_products) AS n_products, MIN(order_date) AS order_datetime, ROUND(SUM(item_quant_cost),2) AS order_cost 
  FROM r2o_cost_view
  GROUP BY order_id
  ORDER BY order_cost DESC
  """)
r2oc_agg.show()
```
